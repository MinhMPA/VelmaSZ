```python
import matplotlib
matplotlib.style.use('Minh_paper_style')
import matplotlib.pyplot as plt
from matplotlib.ticker import (FixedLocator, MultipleLocator, FixedFormatter, FormatStrFormatter,
                               AutoMinorLocator)
import numpy as np
```

```python
filename='/Users/CaMap/lambdas_AAP0.9_085E14_Rozo_maxBCGclusters_singlesampling_specz.txt'
lambdas=np.loadtxt(filename,usecols=0)
plt.subplot(1,1,1)
plt.axhline(0.,ls=':',c='k')
realizations=np.arange(0,len(lambdas))
plt.plot(realizations,lambdas,color='navy')
plt.xlabel(r"Velocity realizations", fontsize=20)
plt.ylabel(r"Weight", fontsize=20)
```

```python
filename='/Users/CaMap/Desktop/table_for_minh.txt'
r=np.loadtxt(filename,usecols=0)
a=np.loadtxt(filename,usecols=1)
err=np.loadtxt(filename,usecols=2)
fig, ax = plt.subplots()
singh_a=4.6
singh_err=0.5
ax.axhline(0.,color='k',ls='--')
ax.plot(r,a,color='royalblue', label='this work')
ax.fill_between(r,a-err,a+err,color='royalblue',alpha=0.3)
ax.axhline(singh_a,ls=':',c='orangered',label='Singh et al. 2015')
ax.fill_between(r,singh_a-singh_err,singh_a+singh_err,color='orangered',alpha=0.2)
ax.set_xlabel(r"Filter scale", fontsize=20)
ax.set_ylabel(r"IA amplitude", fontsize=20)
ax.set_xlim(0.,60.)
ax.set_ylim(-5.,10.)
ax.yaxis.set_major_locator(MultipleLocator(2.))
ax.yaxis.set_minor_locator(MultipleLocator(1.))
ax.xaxis.set_major_locator(MultipleLocator(20.))
ax.xaxis.set_minor_locator(MultipleLocator(10.))
ax.legend(frameon=False, fontsize=18)
```

```python
alpha = np.random.normal(loc=0.9,scale=0.18,size=837)
fig, ax = plt.subplots()
bins=np.linspace(0.,1.8,60)
ax.hist(alpha,bins=bins,color='gold')
ax.set_xlabel(r"Gas fraction w.r.t. cosmic mean", fontsize=20)
ax.set_ylabel(r"Sample count", fontsize=20)
ax.set_xlim(0.,1.85)
ax.set_ylim(0.,80.)
ax.yaxis.set_major_locator(MultipleLocator(20))
ax.yaxis.set_minor_locator(MultipleLocator(10))
ax.xaxis.set_major_locator(MultipleLocator(0.2))
ax.xaxis.set_minor_locator(MultipleLocator(0.1))
ax.axvline(x=alpha.mean(),ls='--',color='navy')
```
